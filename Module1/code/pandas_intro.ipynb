Wait, what just happened? Why did we leave out the parentheses in the "shape" function? The answer is because Python is a general programming language. Programming languages have methods (actions) and attributes (characteristics). Many programming languages use parentheses to denote methods. The programmers of pandas have decided that describe is an action (hence, a method) but shape is a characteristic of a dataset (hence, an attribute). ## Data Visualization 

```python
#histogram of Income using 9 bins
n, bins, patches = plt.hist(vp.Income, 9, facecolor='green', alpha=0.75)
plt.ylim(0, 10)
plt.title('Distribution of Income')
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
#histogram of Age using 7 bins
n, bins, patches = plt.hist(vp.Age, 7, facecolor='blue', alpha=0.75)
plt.ylim(0, 15)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title('Histogram of Age using 7 bins')
plt.show()
```

```python
plt.scatter(vp.Income,vp.Age)
plt.title('Scatter of Income and Age')
plt.xlabel("Income")
plt.ylabel("Age")
plt.xlim(0, 100000)
plt.show()
```

```python
#crosstab
video_cross=pd.crosstab(vp.Gender, vp.Genre,colnames=['Genre'],rownames=['Gender'])
print(video_cross)
```

```python
#bar chart of Genre by Gender
video_cross.plot(kind='bar', stacked=True)
plt.title('Movie Genre by Gender')
```
